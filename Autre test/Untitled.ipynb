## Importer les Bibliothèques

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,  mean_squared_error
```

## Utilise Pandas pour charger les données

```python
path = r"C:\Users\21265\OneDrive - Ecole Centrale Casablanca\INFORMATIQUE\Machine learning\Autre test\data .txt"
D = pd.read_table(path)
```

## Préparation des données
### Y c'est la colonne PAC et X c'est le reste du tableau sans la colonne PAC
### Créer un modèle de regression linéaire qui va prédire les valeur de Y sachant X 


```python
Y = D.PAC
X = D.drop('PAC', axis = 1) 
```

## Diviser les données en training set et test set avec 2348 éléments dans le test

```python
XTrain, XTest, yTrain, yTest = train_test_split(X,Y, test_size = 2348, random_state = 100)
```

```python
print("Taille de XTrain :", XTrain.shape[0])
print("Taille de XTest :", XTest.shape[0])
print("Total : ", XTrain.shape[0] + XTest.shape[0])
```

## Création du modèle

```python
model = LinearRegression()
```

## On injecte les données d'entraînemment dans le modèle 

```python
model.fit(XTrain, yTrain)
```

## On regarde la performance du modèle sur les données d'entraînement

```python
model.score(XTrain, yTrain)  ## Le modèle est précis à 83,8% sur l'entraînement
```

## On évalue la performance du modèle sur les données de test (C'est la vraie performance : Accuracy du modèle)

```python
model.score(XTest, yTest)  ## On trouve 84% c'est pas mal
```

```python
print(model.intercept_)
print(model.coef_)
```

## Les graphes

```python
plt.figure(figsize = (20,20))
plt.subplot(2,1,1)
plt.xlabel('Y')
plt.ylabel('Valeur prédite')
plt.scatter(yTrain, model.predict(XTrain), c = 'g', label = 'Y en fonction des valeurs prédites')

plt.legend()
plt.show()

plt.figure(figsize = (20,20))
plt.subplot(2,1,2)
plt.xlabel('Y')
plt.ylabel('résidus')
plt.scatter(yTrain, yTrain - model.predict(XTrain), c = 'r', label= 'Y en fonction des résidus' )
plt.legend()
plt.show()
```

## Erreur quadratique moyenne MSE

```python
## Calcul de l'erreur quadratique moyenne sur les données d'apprentissage
y_pred = model.predict(XTrain)
erreur_quad_Appr = mean_squared_error(yTrain, y_pred)
print("L'erreur quadratique moyenne sur les données d'apprentissage est :{:.2f}".format(erreur_quad_Appr))
```

```python
## Calcul de l'erreur quadratique moyenne sur les données de test
y_pred = model.predict(XTest)
erreur_quad_Test = mean_squared_error(yTest, y_pred)
print("L'erreur quadratique moyenne sur les données de test est: {:.2f} ".format(erreur_quad_Test))
```
