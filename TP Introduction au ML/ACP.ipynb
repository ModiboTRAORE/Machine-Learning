## J'importe les bibliothèques utiles

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

plt.style.use('ggplot')
```

```python
dataset = pd.read_csv(r'Data\meteo.csv')
dataset.head()
```

```python
dataset = dataset[['Total.Precipitation.daily.sum..sfc.',
                 'Total.Cloud.Cover.daily.mean..sfc.', 
                 'High.Cloud.Cover.daily.mean..high.cld.lay.',
                 'Medium.Cloud.Cover.daily.mean..mid.cld.lay.',
                 'Low.Cloud.Cover.daily.mean..low.cld.lay.',
                 'Wind.Direction.daily.mean..900.mb.',
                 'Total.Cloud.Cover.daily.max..sfc.',
                 'High.Cloud.Cover.daily.max..high.cld.lay.',
                 'Medium.Cloud.Cover.daily.max..mid.cld.lay.',
                 'Low.Cloud.Cover.daily.max..low.cld.lay.',
                 'Wind.Speed.daily.max..10.m.above.gnd.',
                 'Wind.Speed.daily.max..80.m.above.gnd.',
                 'Wind.Speed.daily.max..900.mb.',
                 'Wind.Gust.daily.max..sfc.',
                ]]
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(dataset)
x_scaled
```

## K-Means

```python
# Somme des distances

distance_clusters = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++')
    kmeans.fit(x_scaled)
    distance_clusters.append(kmeans.inertia_)
```

```python
distance_clusters
```

```python
plt.plot(range(1,11), distance_clusters)
plt.title('La méthode du coude')
plt.xlabel('Nombre de clusters')
plt.ylabel('Somme des distances')
plt.show()
```

```python
# Les deux premiers composantes
kmeans = KMeans(n_clusters =2, init = 'k-means++')
```

```python
kmeans.fit(x_scaled)
```

```python
clusters = kmeans.predict(x_scaled)
```

```python
#j'ajoute une colonne cluster à la dataset
dataset['cluster'] = clusters
dataset.head()
```

```python
dataset_cluster = dataset.groupby(['cluster']).median().round(0)

dataset_cluster
```

```python
## Fonction pour afficher les composantes principales

def display_scree_plot(pca):
    scree = pca.explained_variance_ratio_*100
    plt.bar(np.arange(len(scree))+1, scree)
    plt.plot(np.arange(len(scree))+1, scree.cumsum(), c='red', marker = 'o')
    plt.xlabel("rang de l'axe d'inertie")
    plt.ylabel("Pourcentage d'inertie")
    plt.show(block = False)
    
```

```python
n_comp = 2

pca = PCA(n_components = n_comp)
pca.fit(x_scaled)
print('Nombre de composants calculées : ', pca.n_components_)

plt.figure(figsize = (15, 8))
display_scree_plot(pca)
```

```python
Xk_projected = pca.transform(x_scaled)
```

```python

#Projection sur les deux composentes
sns.scatterplot(data = Xk_projected, x = Xk_projected[:,0], y=Xk_projected[:,1], alpha =1 )
```

```python
plt.figure(figsize=(25,20))

# Tracer les variables initiales sur le plan des composantes principales
feature_vectors = pca.components_.T 
for i, v in enumerate(feature_vectors):
    plt.arrow(0, 0, v[0], v[1], color='b', alpha=0.5, width=0.001)
    plt.text(v[0]*1.5, v[1]*1.5, f'Variable {i+1}', color='black')
plt.xlabel('Première composante principale')
plt.ylabel('Deuxième composante principale')
plt.title('Plan des composantes principales')
plt.grid(True)

plt.show()
```
