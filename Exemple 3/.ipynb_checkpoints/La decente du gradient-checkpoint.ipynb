## Regression linéaire

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression  #permet de générer une base de données
```

```python
x, y = make_regression(n_samples = 100, n_features = 1, noise =10) # on génère une data de 100 ligne à une variable x
plt.scatter(x, y)
```

```python
print(x.shape)
print(y.shape)
```

```python
y = y.reshape(y.shape[0],1)
print(y.shape)
```

# $Y = X\cdot\theta$  en termes de matrices

On définit la matice $X$ et le vecteur $\theta$ (Aléatoire) un vecteur de dimension (2,1)

```python
X = np.hstack((x, np.ones(x.shape)))
theta = np.random.randn(2,1)
theta
```

## On crée le modèle $F = X\cdot \theta$

```python
def model(X, theta):
    return X.dot(theta)
```

```python
plt.plot(x,model(X, theta), c = 'g')
plt.scatter(x,y)
```

## Calculons la fonction côut $J(\theta) = \frac{1}{2m} \sum (X\cdot \theta - Y)^2$

```python
def fonction_cout(X, theta, y):
    m = len(y)
    return 1/(2*m) * np.sum((model(X,theta) - y)**2)
```

```python
fonction_cout(X, theta, y)
```

## Gradient $\frac{\partial J}{\partial \theta} = \frac{1}{m} X^T(X\cdot \theta -Y)$ Descente de gradient $\theta = \theta - \alpha \frac{\partial J}{\partial \theta}$

```python
def grad(X, y , theta):
    m = len(y)
    return 1/m * X.T.dot(model(X,theta) - y)
```

```python
def descente_gradient(X, Y, theta, learning_rate, n_iterations):
    for i in range(0, n_iterations):
        theta = theta - learning_rate*grad(X, y , theta)
    return theta
```

```python
theta_final = descente_gradient(X, y, theta, learning_rate = 0.01, n_iterations = 1000)
theta_final
```

```python
predictions = model(X, theta_final)
plt.scatter(x, y)
plt.plot(x, predictions, c= 'r')
```
